```python
def calculate(exp):
	if '+' in exp:
		exp = exp.split('+')
		ans = int(exp[0]) + int(exp[1])
		label = tk.Label(frame, text=f"The answer is: {ans}", bg="#e8de1a")
		label.grid(row=2, column=0, columnspan=2, pady=5)
	elif '-' in exp:
		exp = exp.split('-')
		ans = int(exp[0]) - int(exp[1])
		label = tk.Label(frame, text=f"The answer is: {ans}", bg="#e8de1a")
		label.grid(row=2, column=0, columnspan=2, pady=5)
	elif '*' in exp:
		exp = exp.split('*')
		ans = int(exp[0]) * int(exp[1])
		label = tk.Label(frame, text=f"The answer is: {ans}", bg="#e8de1a")
		label.grid(row=2, column=0, columnspan=2, pady=5)
	elif '/' in exp:
		exp = exp.split('/')
		ans = int(exp[0]) / int(exp[1])
		label = tk.Label(frame, text=f"The answer is: {ans}", bg="#e8de1a")
		label.grid(row=2, column=0, columnspan=2, pady=5)
		

def get_info_and_destroy():
        exp = name_entry.get()
        
        calculate(exp) 

root = tk.Tk()
root.title("Calculator")

root.geometry("480x240") 
root.config(bg='#607B9E')

frame = tk.Frame(root, bg="#333333")
frame.pack(expand=True)

name_label = tk.Label(frame, text="Expression", width=15, height=1, bg="#e8de1a")
name_label.grid(row=0, column=0, sticky="w", padx=(10,5), pady=(10,5))

name_entry = tk.Entry(frame, bg="#e8de1a")
name_entry.grid(row=0, column=1, padx=(5,10), pady=(10,5))  

submit_button = tk.Button(frame, text="calculate", command=get_info_and_destroy, bg="#e8de1a", height=1, width=34)
submit_button.grid(row=1, columnspan=2, pady=5) 

root.mainloop()  
```

```python
def calculate():
    global exp
    if '+' in exp:
        exp = exp.split('+')
        ans = int(exp[0]) + int(exp[1])
        label.config(text=f"{ans}")
    elif '-' in exp:
        exp = exp.split('-')
        ans = int(exp[0]) - int(exp[1])
        label.config(text=f"{ans}")
    elif '*' in exp:
        exp = exp.split('*')
        ans = int(exp[0]) * int(exp[1])
        label.config(text=f"{ans}")
    elif '/' in exp:
        exp = exp.split('/')
        ans = int(exp[0]) / int(exp[1])
        label.config(text=f"{ans}")
def f1():
    global exp
    exp = exp + "1"
    label.config(text=f"{exp}")
def f2():
    global exp
    exp = exp + "2"
    label.config(text=f"{exp}")
def f3():
    global exp
    exp = exp + "3"
    label.config(text=f"{exp}")
def f4():
    global exp
    exp = exp + "4"
    label.config(text=f"{exp}")
def f5():
    global exp
    exp = exp + "5"
    label.config(text=f"{exp}")
def f6():
    global exp
    exp = exp + "6"
    label.config(text=f"{exp}")
def f7():
    global exp
    exp = exp + "7"
    label.config(text=f"{exp}")
def f8():
    global exp
    exp = exp + "8"
    label.config(text=f"{exp}")
def f9():
    global exp
    exp = exp + "9"
    label.config(text=f"{exp}")
def f0():
    global exp
    exp = exp + "0"
    label.config(text=f"{exp}")
def fadd():
    global exp
    exp = exp + "+"
    label.config(text=f"{exp}")
def fsub():
    global exp
    exp = exp + "-"
    label.config(text=f"{exp}")
def fmul():
    global exp
    exp = exp + "*"
    label.config(text=f"{exp}")
def fdiv():
    global exp
    exp = exp + "/"
    label.config(text=f"{exp}")
def fback():
    global exp
    exp = exp[:-1]
    label.config(text=f"{exp}")
def clear():
    global exp
    exp = ""
    label.config(text=exp)
root = tk.Tk()
root.title("Calculator")

exp = ""

root.geometry("1080x720") 
root.config(bg='#607B9E')

frame = tk.Frame(root, bg="#333333")
frame.pack(expand=True)

b1 = tk.Button(frame, text="1", command=f1, bg="#e8de1a", height=1, width=34)
b1.grid(row=0, column=0, pady=5, padx=5) 
b2 = tk.Button(frame, text="2", command=f2, bg="#e8de1a", height=1, width=34)
b2.grid(row=0, column=1, pady=5, padx=5) 
b3 = tk.Button(frame, text="3", command=f3, bg="#e8de1a", height=1, width=34)
b3.grid(row=0, column=2, pady=5, padx=5) 
b4 = tk.Button(frame, text="4", command=f4, bg="#e8de1a", height=1, width=34)
b4.grid(row=1, column=0, pady=5, padx=5) 
b5 = tk.Button(frame, text="5", command=f5, bg="#e8de1a", height=1, width=34)
b5.grid(row=1, column=1, pady=5, padx=5) 
b6 = tk.Button(frame, text="6", command=f6, bg="#e8de1a", height=1, width=34)
b6.grid(row=1, column=2, pady=5, padx=5) 
b7 = tk.Button(frame, text="7", command=f7, bg="#e8de1a", height=1, width=34)
b7.grid(row=2, column=0, pady=5, padx=5) 
b8 = tk.Button(frame, text="8", command=f8, bg="#e8de1a", height=1, width=34)
b8.grid(row=2, column=1, pady=5, padx=5) 
b9 = tk.Button(frame, text="9", command=f9, bg="#e8de1a", height=1, width=34)
b9.grid(row=2, column=2, pady=5, padx=5) 
b0 = tk.Button(frame, text="0", command=f0, bg="#e8de1a", height=1, width=34)
b0.grid(row=3, column=0, pady=5, padx=5) 
badd = tk.Button(frame, text="+", command=fadd, bg="#e8de1a", height=1, width=34)
badd.grid(row=3, column=1, pady=5, padx=5) 
bsub = tk.Button(frame, text="-", command=fsub, bg="#e8de1a", height=1, width=34)
bsub.grid(row=3, column=2, pady=5, padx=5) 
bmul = tk.Button(frame, text="*", command=fmul, bg="#e8de1a", height=1, width=34)
bmul.grid(row=4, column=0, pady=5, padx=5) 
bdiv = tk.Button(frame, text="/", command=fdiv, bg="#e8de1a", height=1, width=34)
bdiv.grid(row=4, column=1, pady=5, padx=5)
bback = tk.Button(frame, text="<---", command=fback, bg="#e8de1a", height=1, width=34)
bback.grid(row=4, column=2, pady=5, padx=5)
bback = tk.Button(frame, text="CLEAR", command=clear, bg="#e8de1a", height=1, width=34)
bback.grid(row=5, columnspan=3, pady=5, padx=5)
label = tk.Label(frame, text=f"{exp}", bg="#e8de1a", width=75)
label.grid(row=6,columnspan=3, pady=5, padx=5)
submit_button = tk.Button(frame, text="Calculate", command=calculate, bg="#e8de1a", height=1, width=34)
submit_button.grid(row=7, columnspan=3, pady=5, padx=5) 
root.mainloop() 
```
